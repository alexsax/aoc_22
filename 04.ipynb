```python
import numpy as np
def range_subset(range1, range2):
    """Whether range1 is a subset of range2."""
    if not range1:
        return True  # empty range is subset of anything
    if not range2:
        return False  # non-empty range can't be subset of empty range
    if len(range1) > 1 and range1.step % range2.step:
        return False  # must have a single value or integer multiple step
    return range1.start in range2 and range1[-1] in range2

def get_range(vals):
    s, e = vals.split('-')
    return range(int(s), int(e)+1)

pairs_split = [pair.split(',') for pair in pairs.split('\n')]
pairs_split = [(get_range(first), get_range(second)) for first, second in pairs_split]
vals = [range_subset(first, second) or range_subset(second, first) for first, second in pairs_split]
print(sum(vals))
```
